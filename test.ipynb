```python
terms1 = ["people", "closer"]
terms2 = ["first", "look"]
terms3 = ["article", "first"] 

docs = ["ago cost ever hardly lines look new oakleys pair wore year",
        "article che closer look people polls send take writes",
        "arrogance christians hsu like look melinda mhsu share thoughts topic writes",
        "article dwight experts first look note tuinstra tuinstra writes",
        "access files found look new public system",
        "anyone claim compare look new price taken terms viewsonic",
        "come creed dead life look resurrection world",
        "companies compatible computer following gave less look monitors monochrome pc quick sell shopper",
        "back bike drom get look proceding today work writes",
        "anyone could first happen idea isolated know know look morphine newsgroup opium",
        "article closer look people polls send take writes",
        "bitmap bitmap colours data data exactly format height like look mean stored width windows",
        "abc appearance coverage csa excellent excellent icop look loved modern originator production sleek",
        "across application come framework good hello inmark investigating look object oriented purchase zapp zinc",
        "avoid crash crashing fails faq headers look machine never newsreader restarting sheet way",
        "code considering could find first like like look sample something source widget widgets writing",
        "closer deep look njd people polls send take team",
        "build buy devices devices electronics else fails look mags make popular radio",
        "color could current get higher lines look patches please ro someone tell xterm",
        "across came catalogues components components cooling couple effect heat intended look peltier pumps range"]
        
import math
def PPMI(terms, docs=docs):
  w0, w1 = terms
  N = len(docs)
  
  def count2(w0, w1):
    count = 0
    for doc in docs:
      if w0 in doc.split() and w1 in doc.split():
        count += 1
    return count
  
  def count1(w):
    count = 0
    for doc in docs:
      if w in doc.split():
        count += 1
    return count

  return max(math.log(N * count2(w0, w1) / (count1(w0) * count1(w1)), 2),0)

print(f"Для термов {terms1[0]} и {terms1[1]}, PPMI = %.3f" %PPMI(terms1))
print(f"Для термов {terms2[0]} и {terms2[1]}, PPMI = %.3f" %PPMI(terms2))
print(f"Для термов {terms3[0]} и {terms3[1]}, PPMI = %.3f" %PPMI(terms3))
```

```python

test_size = 0.25
random_state = 65
C = 1.5


import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/tweets.csv')
print(df)
df = df[["text", "target"]]
#df = df.drop_duplicates("text")

import nltk
from nltk import stem
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import precision_recall_fscore_support
import re

nltk.download("stopwords", quiet=True)
nltk.download('punkt', quiet=True)
stemmer = SnowballStemmer("english")
stopwords = set(stopwords.words('english'))


def preprocess(text):
  text = re.sub(r'[^\w\s]', '', text.lower())
  text = text.lower()
  text_tokens = text.split(" ")
  tokens_without_sw = [stemmer.stem(word) for word in text_tokens if not word in stopwords]
  return (" ").join(tokens_without_sw)

df['text'] = df['text'].apply(preprocess)

y = df['target'].values
X_train, X_test, y_train, y_test = train_test_split(df['text'], y, test_size=test_size, random_state=random_state)
vectorizer = TfidfVectorizer(decode_error='ignore')
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

model = LinearSVC(random_state = random_state, C = C).fit(X_train, y_train)
predictions = model.predict(X_test)

precision, recall, f1score, support = precision_recall_fscore_support(y_test, predictions)
print("Precision (macro avg): %.3f" %(sum(precision)/len(precision)))
print("Recall (macro avg): %.3f" %(sum(recall)/2))
print("F-score (macro avg): %.3f" %(sum(f1score)/2))
```
